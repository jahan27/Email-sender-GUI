```python
def send():
    to = to_entry.get()
    from_ = from_entry.get()
    subject = subject_entry.get()
    content= email.get(1.0, tk.END)
    
    #checking
    
    first_ltr=to[0]
    first_ltr_from=from_[0]
    
    
    if to=='' or from_==''or content=='':
            error_window= messagebox.showerror(message="Error!")
    
    elif '@' not in to or '.' not in to or '@' not in from_ or '.' not in from_:
            error_window= messagebox.showerror(message="Error! Email is invalid")
    
    elif first_ltr.isalnum() == False or first_ltr_from.isalnum() == False:
        error_window= messagebox.showerror(message="Error! number cannot start so it is invalid")
    
    
    fullemail= EmailMessage()
    fullemail['Subject']= subject
    fullemail['To']= to
    fullemail['From']= from_
    fullemail.set_content(content)

    try:
        server= smtplib.SMTP_SSL("smtp.gmail.com",465)
        server.login(from_,"password")
        server.send_message(fullemail)
        server.quit()
        
    except:
        error_window= messagebox.showerror(message="Error! Something went wrong")

    
    
    
    
    
import tkinter as tk
from tkinter import messagebox 
import smtplib
from email.message import EmailMessage 
    
    
window=tk.Tk()
window.geometry("800x500")
window.title("GUI Tutorial")

info_frame= tk.Frame(window, bg="#ff0000")
info_frame.place(relx=0,rely=0,relwidth=1,relheight=0.3)

email_frame= tk.Frame(window, bg="#00ff00")
email_frame.place(relx=0,rely=0.3,relwidth=1,relheight=0.6)

attach_frame= tk.Frame(window, bg="#0000ff")
attach_frame.place(relx=0,rely=0.9,relwidth=1,relheight=0.1)

#-----------info---------------
label1= tk.Label(info_frame,text="To:", font=("Arial",12), fg="#000000")
label1.place(relx=0.05,rely=0.2)

label2= tk.Label(info_frame,text="From:", font=("Arial",12), fg="#000000")
label2.place(relx=0.05,rely=0.4)

label3= tk.Label(info_frame,text="Subject:", font=("Arial",12), fg="#000000")
label3.place(relx=0.05,rely=0.6)

to_entry= tk.Entry(info_frame, bd=2, font=("Arial",10))
to_entry.place(relx=0.15,rely=0.2,relwidth=0.25, relheight=0.16)

from_entry= tk.Entry(info_frame, bd=2, font=("Arial",10))
from_entry.place(relx=0.15,rely=0.4,relwidth=0.25, relheight=0.16)

subject_entry= tk.Entry(info_frame, bd=2, font=("Arial",10))
subject_entry.place(relx=0.15,rely=0.6,relwidth=0.25, relheight=0.16)

#-----------email--------------

email= tk.Text(email_frame,bg= "#ffffff", fg="#000000",font=("Arial", 12), padx=10, pady=10)
email.place(relx=0,rely=0, relheight=1, relwidth=1)



#---------attach----------------

send_button= tk.Button(attach_frame, text="send", activeforeground="#ffffff", command= send)
send_button.place(relx=0.75,rely=0.2,relheight=0.6,relwidth=0.2)















window.mainloop()
```
